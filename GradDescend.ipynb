```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
```

```python
def gradient_descend(x,y):
    mi = 0
    ci = 0
    lr = 0.00001
    n = len(x)
    iteration = 1000
    cost1 = 0
    plt.scatter(x,y)
    for i in range(iteration):
        y_pred = mi*x + ci
        cost = (1/n)*sum([value**2 for value in (y-y_pred)])
        md = -(2/n)*sum(x*(y-y_pred))
        cd = -(2/n)*sum(y-y_pred)
        mi = mi - lr * md
        ci = ci - lr * cd
        plt.plot(x,y_pred,color = 'red')
        if math.isclose(cost, cost1, rel_tol=1e-20):
            break
        cost1 = cost
        print ("m {}, b {}, md {}, cd {}, cost {}, iteration {}".format(mi,ci,md,cd,cost, i))
    
    return mi, ci
```

```python
var = pd.read_csv('test_Scores.csv')
var
```

```python
x = var.math
y = var.cs
var.cs.sum()
```

```python
gradient_descend(x,y)
```
